#### Data cleaning - Dropping the missing values records and ensuring we have correct data types

```python
# Check for missing values
print(df.isnull().sum())

# Fill or drop missing values
df = df.dropna()  

# Ensure correct data types
df['County_FIPS_Code'] = df['County_FIPS_Code'].astype(int)
df['StudyYear'] = df['StudyYear'].astype(int)
```

#### Exploratory Data Analysis (EDA)

#### Summary statistics for median household income across states

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics for Median Household Income by State 
state_mhi_summary = df.groupby('State_Name')['MHI'].describe() 
print(state_mhi_summary)
```

#### Distribution of Median Household Income by State

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot of Median Household Income by State
plt.figure(figsize=(14, 8))
sns.boxplot(x='State_Name', y='MHI', data=df, palette='muted')
plt.title('Distribution of Median Household Income by State')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Median Household Income (in dollars)')
plt.show()
```

#### Average Median Household Income by State

```python
# Average Median Household Income by State
average_mhi = df.groupby('State_Name')['MHI'].mean()

# Plot
plt.figure(figsize=(14, 8))
average_mhi.plot(kind='bar', color='skyblue')
plt.title('Average Median Household Income by State')
plt.ylabel('Average Median Household Income (in dollars)')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.show()
```

#### Exploring Trends Over Time - median household income has changed over time for each state.

```python
# Ensure StudyYear is in datetime format
df['StudyYear'] = pd.to_datetime(df['StudyYear'], format='%Y')

# Group by year and state, then calculate the mean Median Household Income
income_trend_by_state = df.groupby([df['StudyYear'].dt.year, 'State_Name'])['MHI'].mean().unstack()

# Plot
plt.figure(figsize=(14, 8))
income_trend_by_state.plot(kind='line', figsize=(14, 8))
plt.title('Trend of Median Household Income Over Years by State')
plt.ylabel('Median Household Income (in dollars)')
plt.xlabel('Year')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

#### Trends in Median Household Income

Evaluate how median household income has changed over the years.

#### Cost of Center-Based vs. Family Childcare

Comparing the median prices of center-based care and family childcare for different age groups

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Median prices of center-based care
center_based_columns = ['MCBto5', 'MC6to11', 'MC12to17', 'MC18to23', 'MC24to29', 'MC30to35', 'MC36to41', 'MC42to47', 'MC48to53', 'MC54toSA']
center_based_prices = df[center_based_columns].mean()

# Median prices of family childcare
family_childcare_columns = ['MFCCBto5', 'MFCC6to11', 'MFCC12to17', 'MFCC18to23', 'MFCC24to29', 'MFCC30to35', 'MFCC36to41', 'MFCC42to47', 'MFCC48to53', 'MFCC54toSA']
family_childcare_prices = df[family_childcare_columns].mean()

# Plot
plt.figure(figsize=(14, 8))
plt.bar(center_based_columns, center_based_prices, color='blue', alpha=0.6, label='Center-Based Care')
plt.bar(family_childcare_columns, family_childcare_prices, color='orange', alpha=0.6, label='Family Childcare', bottom=center_based_prices)
plt.xticks(rotation=45)
plt.title('Median Prices of Center-Based Care vs. Family Childcare Across Age Groups')
plt.ylabel('Median Price (in dollars)')
plt.xlabel('Age Groups')
plt.legend()
plt.show()
```
