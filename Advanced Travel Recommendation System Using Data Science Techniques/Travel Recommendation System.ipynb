```python
import pandas as pd
```

```python
# Step 1. Exported all the dataframes - flat file, web and api data through pickle and are read in this notebook
```

```python
# dataframe for flat file data
df_flat_file = pd.read_pickle('df.pkl_flat_file_data')
```

```python
# dataframe for website data
df_web = pd.read_pickle('df.pkl_web_data')
```

```python
# dataframe for api data
df_api = pd.read_pickle('df.pkl_api_data')
```

```python
df_api.head()
```

```python
# Step 2. Using SQLLITE3 we are loading all these 3 dataframes to 3 different databases
```

```python
import sqlite3
```

```python
conn_flat_file = sqlite3.connect('my_flat_file_db.db')
```

```python
conn_web_data = sqlite3.connect('my_web_data_db.db')
```

```python
conn_api_data = sqlite3.connect('my_api_data_db.db')
```

```python
# flat file table database creation and data load
flat_file_table = 'flat_file_table_db'  # table name
df_flat_file.to_sql(flat_file_table, conn_flat_file, if_exists='replace', index=False)
```

```python
# web table database creation and data load
web_data_table = 'web_data_table_db'  # table name
df_web.to_sql(web_data_table, conn_web_data, if_exists='replace', index=False)
```

```python
# api table database creation and data load
api_data_table = 'api_data_table_db'  # table name
df_api.to_sql(api_data_table, conn_api_data, if_exists='replace', index=False)
```

```python
query = f'SELECT * FROM {flat_file_table}'
result_flat_file_df = pd.read_sql(query, conn_flat_file)
print(result_flat_file_df)
```

```python
query = f'SELECT * FROM {web_data_table}'
result_web_data_df = pd.read_sql(query, conn_web_data)
print(result_web_data_df)
```

```python
query = f'SELECT * FROM {api_data_table}'
result_api_data_df = pd.read_sql(query, conn_api_data)
print(result_api_data_df)
```

```python
# Step 3. Merging all the databases into a single merged table - merged_table
```

```python
# Load data from SQLite tables into DataFrames

conn = sqlite3.connect('my_merged_database.db')

# Merge DataFrames based on a common key (City)
merged_df = pd.merge(result_flat_file_df, result_web_data_df, on='City', how='inner')
merged_df = pd.merge(merged_df, result_api_data_df, on='City', how='inner')

# Now 'merged_df' contains the combined data from all three datasets
```

```python
merged_table = 'my_merged_database'  # table name
merged_df.to_sql(merged_table, conn, if_exists='replace', index=False)
```

```python
merged_df.head()
```

```python
# reviewing all the columns after merged table
for col in merged_df.columns:
    print(col)
```

```python
query = f'SELECT * FROM {merged_table}'
result_merged_df = pd.read_sql(query, conn)
print(result_merged_df)
```

```python
# Step 4. Using merged table we are creating below listed virtualization charts
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from dateutil import parser
from matplotlib import style
import seaborn as sns  # For better styling
```

```python
# Virtualization - 1 (line plot chart) - for City and Number of visitors in millions 
```

```python
query = f'SELECT City, avg(Visitors) FROM {merged_table} GROUP BY City'
result_query_df = pd.read_sql(query, conn)
print(result_query_df)
```

```python
# Line Plot

sns.lineplot(x='City', y='avg(Visitors)', data=result_query_df)

plt.title("City and Average Visitors_(millions)", fontsize=15, color="Black")
plt.xlabel('City')
plt.ylabel('Visitors_(millions)')
plt.grid(False)
plt.tight_layout()

plt.show()
```

```python
# Virtualization - 2 (bar chart) - for Tourist attraction and Number of visitors in millions 
```

```python
query = f'SELECT Tourist_attraction, City, Visitors FROM {merged_table} GROUP BY Tourist_attraction'
result_query_df = pd.read_sql(query, conn)
print(result_query_df)
```

```python
# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(result_query_df['Tourist_attraction'], result_query_df['Visitors'], label='in Millions', color='skyblue', alpha=0.7)

plt.xlabel('Tourist Attractions')
plt.ylabel('Number of Visitors')
plt.title('Visitors to Tourist Attractions')
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

```python
# Virtualization - 3 (box plot) - for City and Average Hotel review rating in that city
```

```python
query = f'SELECT City, Name_x, avg(ReviewRating) FROM {merged_table} GROUP BY City, Name_x'
result_query_df = pd.read_sql(query, conn)
print(result_query_df)
```

```python
# Create the box plot
sns.set(style="whitegrid")  
plt.figure(figsize=(8, 6))
sns.boxplot(x='City', y='avg(ReviewRating)', data=result_query_df)
plt.xlabel('Hotels in City')
plt.ylabel('Avg. Hotel Review Rating')
plt.title('Review Ratings of Hotels by Cities')
plt.xticks(rotation=45)
plt.show()
```

```python
# Virtualization - 4 (histogram plot) - for City and Number of hotels in that city
```

```python
query = f'SELECT Name_x, City, avg(ReviewRating) FROM {merged_table} GROUP BY Name_x'
result_query_df = pd.read_sql(query, conn)
print(result_query_df)
```

```python
city_counts = result_query_df['City'].value_counts()
plt.bar(city_counts.index, city_counts.values, color='green', alpha=0.7)
plt.xlabel('City')
plt.ylabel('Number of Hotels')
plt.title('Hotel Distribution by City')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
# Virtualization - 5 (scatter plot) - for City and Hotel review rating in that city
```

```python
plt.scatter(result_query_df['City'], result_query_df['avg(ReviewRating)'], marker='o', color='purple', alpha=0.7)
plt.xlabel('City')
plt.ylabel('Rating')
plt.title('Hotel Ratings by City')
plt.grid(False)
plt.show()
```
